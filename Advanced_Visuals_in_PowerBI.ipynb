**Question 1: What is the difference between a Basic Filter, an Advanced Filter, and a Slicer in Power BI?**
- Basic filter in Power BI is a filter that allows data filtering directly by selecting specific values from multiple fields. This filter is available while creating the report, but not available to the end user.
- On the other hand, an advanced filter allows complexity in drawing data based on complex conditions using multiple sets of rules and operators such as equals, contains, greater than, less than, and between. It is also applied through the Filters pane. This filter is also used while developing the report.
- On the slicer is a visual filter that is embedded in the report, which makes the report more interactive by giving users the option to select the options with respect to which data is required.

**Question 4: What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?**

**Visual Level Filter:** In this case, the filter is applied to the specific visual that has been selected. For example, if a pie chart is selected, the filter will be applied to that specific visual only and the rest of the report or other pages do not gets effected.

**Page-level filter:** This filter is applied to the whole page, not the complete report. Suppose there are 5 pages in a report; the filter will be applied to that specific page only, and all 5 pages.

**Report Level Filter:** In this case, the complete report shows the data based on the filters applied. All the filters are applied to all visuals under all pages. Except for some specific visuals that have been calculated to show specific values.

**Question 10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.**

- Drill allows us to move through hierarchical levels within the same visual. In this, we start broad and progressively uncover details. For example, suppose a sales report is presented, so in drill down it may first show the year-wise sales, then if we go deep it will show quarterly, then monthly.

- On the other hand, in drill through, we are taken from a summarised page to a detailed page. When we go deep, the next page may show details such as product-wise sales based on region division and month-wise sales. So this present more detailed report.
