**Question 1: A die is rolled. What is the probability of getting:**

**(a) An even number**

    P (even number) = Number of favourable outcomes / Total number of possible outcomes
    = 3/6
    = ½ or 0.5
**(b) A number greater than 4**

    P (A number greater than 4) = Number of favourable outcomes / Total number of possible outcomes
    = 2/6
    = 1/3 or 0.3333


**Question 2: In a class of 50 students:**

**20 like Mathematics (M)**

**15 like Science (S)**

**5 like both subjects**

**What is the probability that a student chosen at random likes Mathematics or Science?**

    P (Mathematics or Science) = Number of favourable outcomes / Total number of possible outcomes
    = 30/50
    = 3/5 or 0.6

Students who only like maths = 20-5 = 15

Students who like only science = 15- 5 = 10

Total number of students who like maths or science = 15 + 10 + 5 = 30

**Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm. Find the sampling error**

    Sampling Error = Population Mean – Sampling Mean
    = 160-158
    = 2 cm

- 2cm is the sampling error


**Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)**

    Standard Error of the Mean (SEM) = Population stdev / squareroot(sample size)
    = 5000 / √100
    = 5000 / 10
    = 500

- Therefore, the sample mean can deviate by 500 from the population mean.

**Question 10: Differentiate between discrete and continuous random variables with examples.**

- **Discrete variable:** Variables that take on countable values. For example, the number of students, number of t-shirts, like 30, 21, 53, 34

- **Continuous variable:** Variables that take on uncountably infinite values within an interval. For example, height of student (162.2 cm, 155.3 cm), temperature (23.3 c, 34.5 c)
