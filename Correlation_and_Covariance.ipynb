**Question 1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.**

- Covariance measures how two variables change together. It is calculated as the average product of deviations of each variable from their mean.
- Correlation is a standardized version of covariance. It rescales covariance to lie between −1 and + 1 .

**Key difference:**
- Covariance depends on the scale of the variables (units matter).
- Correlation removes scale effects, making it unit‑free and easier to interpret.


**Question 2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?**

- Positive covariance: When one variable increases, the other tends to increase too (direct relationship).
- Negative covariance: When one variable increases, the other tends to decrease (inverse relationship).
- Zero covariance: No linear relationship between the variables (though nonlinear relationships may still exist).


**Question 3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?**
- Scale dependence: The magnitude of covariance is affected by the units of measurement, making comparisons across datasets difficult.
- Interpretation difficulty: A large covariance doesn’t necessarily mean a strong relationship it could just reflect large scales of measurement.
- Correlation is preferred because it standardizes covariance, giving a consistent measure of strength and direction of linear association that is comparable across different datasets.


**Question 4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?**

**Pearson’s correlation coefficient:**
- Measures the strength and direction of a linear relationship between two continuous variables.
- Sensitive to outliers and assumes normally distributed data.

**Spearman’s rank correlation coefficient:**
- Based on ranked values, not raw data.
- Captures monotonic relationships (whether increasing or decreasing, not necessarily linear).

**When to use Spearman:**
- Data are ordinal or not normally distributed.
- Relationship is monotonic but not linear.
- Outliers may distort Pearson’s correlation.


**Question 5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?**
- A correlation of 0.85 indicates a strong positive linear relationship between X and Y: as X increases, Y tends to increase significantly.

**Causation cannot be inferred:**
- Correlation only shows association, not cause‑and‑effect.
- The relationship could be due to a third variable (confounding factor) or coincidence.
- To establish causation, controlled experiments or causal inference methods are required.
