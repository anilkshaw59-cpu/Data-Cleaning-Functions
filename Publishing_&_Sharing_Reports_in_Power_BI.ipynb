**Q1. Explain why publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop.**

- Building a report in Power BI Desktop is only the first step. Publishing and sharing make the report available to other users so they can view, interact with, and use it for decision-making. This allows teams to collaborate and access the latest data from anywhere.


**Q2. Define Power BI Service and explain its role in comparison to Power BI Desktop in the reporting lifecycle.**

- Power BI Desktop is used to create reports, import data, and build visualizations. Power BI Service is the online platform where reports are published, shared, refreshed, and managed. Desktop is used for development, while Service is used for collaboration and distribution.


**Q3. What does publishing a report mean in Power BI? List any two outcomes that occur after a report is published.**

Publishing a report means uploading it from Power BI Desktop to Power BI Service.
Two outcomes are:
1) The report becomes available online for viewing.
2) It can be shared with other users or included in dashboards and apps.


**Q4. List and explain the steps involved in publishing a report from Power BI Desktop to Power BI Service in the correct sequence.**

- Create and save the report in Power BI Desktop.
- Sign in to your Power BI account.
- Click the Publish button.
- Select the destination workspace.
- Wait for the upload to be completed.
- Open Power BI Service to view and share the published report.


**Q5. Differentiate between Report and Dashboard in Power BI based on structure and usage.**

**Report**
- Can have multiple pages
- Created in Power BI Desktop
- Contains detailed visuals
- Used for detailed analysis

**Dashboard**
- Has only one page
- Created in Power BI Service
- Shows summary visuals
- Used for quick monitoring


**Q6. Explain the difference between the following three sharing methods in Power BI Service. Also mention one use case for each.**
- Direct Report Sharing: Shares one report with specific users. Sharing a report with a manager.
- Workspace Access: Gives team members access to all content in a workspace based on their role. A BI team working together on reports.
- Power BI Apps: Shares multiple reports and dashboards with many users in read-only mode. Sharing reports with the whole company.


**Q7. Describe the four workspace roles available in Power BI and explain how access control is managed using these roles.**
- Admin: Full control over the workspace.
- Member: Can create, edit, and publish reports.
- Contributor: Can create and edit content but cannot manage the workspace.
- Viewer: Can only view reports.
Access is controlled by assigning the right role to each user based on what they need to do.


**Q8. Why are Power BI Apps considered more secure and scalable than direct sharing?**
- Power BI Apps are easier to manage because one app can be shared with many users. Users only get read-only access, and any updates made to the app are automatically available to everyone. This makes Apps more secure and better for large organizations.


**Q9. Explain Row-Level Security (RLS) in Power BI.**
- Row-Level Security (RLS) limits what data a user can see. Different users open the same report, but each person only sees the data they are allowed to access. For example, a sales manager only sees sales from their own region


**Q10. Assume an organization shares reports without using Apps, workspace roles, or RLS. Discuss any three risks or issues.**
- Users may see confidential data they should not have access to.
- Too many people may get editing permissions, causing accidental changes.
- Managing reports becomes difficult when they are shared individually
