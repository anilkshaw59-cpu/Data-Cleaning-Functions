**Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?**

- It is the default belief; it is assumed that the statement is true until evidence says otherwise. For example, if a company says the average working hours of its bulb is 1000 hours, the default belief will be that it is 1000 hours, unless it is proved otherwise.

- It is important as it provides a baseline assumption against which testing is conducted. It provides a reference point to make a clear judgment on whether to accept or reject the hypothesis.


**Question 2: What does the significance level (α) represent in hypothesis testing?**

- It is the probability of rejecting the hypothesis when it is true. At α = 0.05, we accept a 5% chance of wrongly rejecting H₀.


**Question 3: Differentiate between Type I and Type II errors.**

- Type 1: When the results show the null hypothesis is true and we still rejects it is called a Type 1 error, also known as a false positive.
- Type II Error: When the results show the null hypothesis is false, but we still accept the hypothesis, it is a type II error, also known as a false negative.


**Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each**.

- One-tailed Test: Tests for deviation in one direction only. For example, whether the drug is more effective or not, if it’s the same or worse, the result is the same for them: the drug failed.
- Two-Tailed Test: Tests for deviation in both directions.
For example, if a bulb lasts more than or less than 1000 hours, the null hypothesis is rejected in a two-tailed test hypothesis gets rejected.


**Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.**

- Population Mean = 10 Min
- Sample Mean = 12 Mins
- Sample Size = 9

  T score = xˉ−μ / Standard Error
  = (12-10) / 1
  = 2

  Degree of freedom = n-1
  = 9-1
  = 8

- Critical Value  at significance value 0.05
  = 2.306

- The t-score is less than the critical value, thus the hypothesis cannot be rejected


**Question 6: When should you use a Z-test instead of a t-test**

- Z test
When the sample size is greater than 30, and the population standard deviation is known.
- T test
When the sample size is less than 30, and the population standard deviation is not known.
