**Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.**
**Also mention any two industries where Tableau is commonly used and explain one use case.**

- Tableau is a business intelligence or data visualisation tool that helps in conducting analysis and presenting the analysis in an interactive way for the business decision-making process.
- There are several reasons why Tableau is preferred over excel spreadsheet are:
- Data Handling Capability: Its capability to handle large sets of data. An Excel spreadsheet is good for small data sets, but datasets that contain millions of rows may crash the Excel application.
- Real Time Dashboards: Tableau provides real-time analysis automatically, while in Excel, we have to refresh every time to load new data. Thus, real-time monitoring is possible in Tableau.
- Drag and Drop: In tableau visualization can be made through drag and drop, which makes it more convenient and easier, as compared to excel spread sheet.
- Some of the industries in which Tableau is used are ecommerce industry and the healthcare industry. In the healthcare industry, the data collected through research and patient data can be used for comparative analysis and decision-making. It is also used for monitoring the health of patients.

Q8. **What is the difference between a Worksheet and a Dashboard in Tableau? Why is a dashboard more useful for decision-makers?**

**Worksheet**
- A worksheet is a single view pane where we build one chart or visualization, for example, graphs, a pie chart, a column chart, etc. This particular part is more focused on analyzing a particular aspect of the data. It is mainly used for creating a visualization.

**Dashboard**
- On the other hand dashboard is a collection of multiple worksheets and other objects arranged together that provides a consolidated view of different analyses in one place.

- Dashboard is more powerful for the decision maker because in business, multiple aspect needs to be considered before making any decision. Dashboard provides a more consolidated view of multiple aspects and dimensions. Further, a comparison can be made in the dashboard that is not possible in the worksheet, as it only shows one visual.


**Q9. What is Tableau Prep, and why is it important in real-world data projects?**
Tableau prep is similar to Power Query Editor in Power BI. It is a data preparation tool that allows modification of data, so that it can be made suitable for analysis by removing irregularities.
In the real world, when data is collected from multiple sources, the data can be messy, containing various irregularities, missing data, and incorrect data types. Analysis performed on messy data will give inaccurate results that will hamper the decision making. Tableau Prep ensures the dataset is accurate, consistent, and ready for visualization.

**Then explain how you would:**

**Remove null values:** Tableau Prep → Profile Pane → Identify Null Values → Filter option to exclude null vales.

**Remove duplicate records:** Tableau Prep → Clean Step → Remove Duplicates option

**Change incorrect data types (Answer in steps, no tool access required.):** Tableau Prep → Profile Pane → Adjust Data Type
