**Question 1**

**a) What is data cleaning, and why is it important in data analysis?**  
- Data cleaning is a process through which raw data collected from multiple sources, which is also known as unstructured data. Are transformed into structured data. For example, the standards of data collected from different sources may vary, and it may have errors, blanks and other irregularities. Under the data cleaning process, these irregularities are removed, and the data is made suitable for analysis. For instance, if a column type is marked as text but it only contains integers, then the type is corrected. Blanks and errors in the dataset are corrected under the process.

**b) What are the potential consequences of analysing unclean or messy data?**
- Data is used for decision making purpose; all the insights formed based on data are used for making various decision whether in business, corporations, schools, colleges, hospitals, or other institutions. When uncleaned data is used for analysis, it will compromise the credibility and accuracy of the findings. And the insights derived from such data will not provide an accurate picture, compromising the decision made based on such analysis. Some of the potential consequences of analysing unclean or messy data are as mentioned inaccurate insights, poor decision making, inefficient resource allocation and operational inefficiencies.

**c) Explain the common steps involved in cleaning and organising data.**
- Data cleaning follows a framework called ETL-Extract, Transform and Load. Common steps involved in cleaning and organising data are removing blank spaces, deleting rows with data irregularities, formatting data and columns based on the type of data within the cells, removing duplicates entry and others.

**Question 3. Explain the use of text functions such as TRIM, LEFT, RIGHT, MID, and CONCAT in data cleaning.**

- **TRIM**: Removes all the unnecessary, extra spaces between text, for example, Text Value is Anil   Kumar    Shaw.

         Upon using the TRIM function, the extra two spaces between Anil and Kumar and the extra 3 Spaces between Kumar and Shaw will get removed.

         **Syntax**: TRIM(Cell References or Text)
         **Result**: Anil Kumar Shaw


---


- **LEFT**: Left function provides the exact number of characters from the extreme left side of the text. For example, Cell contains Anil Kumar Shaw

      **Syntax**: =LEFT(Cell Reference or “Anil Kumar Shaw”, 6)
      **Result**: Anil K

Left function also contains space as a character, which is why 6 characters, including space, were provided as a result.


---


- **RIGHT**: Similar to the LEFT function, RIGHT provides a character from the extreme right of the Text.

       **Syntax**: =RIGHT(Cell Reference or “Anil Kumar Shaw”, 6)
       **Result**: r Shaw


---


- **MID**: Similar to LEFT and RIGHT, MID provides a character from in between, in this case start index from the left, and the number of characters has to be provided.

      **Syntax**: =MID(Cell Reference or “Anil Kumar Shaw”, 6, 4)
      **Result**: Kuma

Starting from the sixth character from the left, that is K, 4 Character Kuma was provided as a result.


---


- **CONCAT**: It helps in joining text columns. For instance, Cell A1: First Name: ANIL, Cell B1 Last name: Shaw

         **Syntax**: =CONCAT(A1,B1)
         **Result** = ANILSHAW

ANIL and SHAW joined without a space in between.

**Question 6: Understand and apply fundamental text functions like LEFT, RIGHT, MID, and LEN**.

    =LEFT("ExcelTipsAreGreat", 5) or =LEFT(Cell reference, 5)
    Result: Excel

**Extract the last 4 characters from "DataAnalysis.xlsx" using the RIGHT function**.

    =RIGHT("DataAnalysis.xlsx", 4) or =LEFT(Cell reference, 4)
    Result: xlsx

**Extract the substring "Tips" from "ExcelTipsAreGreat" using the MID function.**

    =MID("ExcelTipsAreGreat", 6, 4) or =MID(Cell Reference, 6,4)
    Result: Tips
**Count the total number of characters in the string "Hello World!" using the LEN function.**

    =LEN("Hello World!")
    Result: 12
**Create a formula to extract the middle 6 characters from "12345-67890-ABCDE".**

    =MID("12345-67890-ABCDE",6,6)
    Result: -67890
