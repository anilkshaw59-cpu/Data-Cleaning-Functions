**Question 1: What is Data Loading in Power BI, and why is it considered the first step of analysis?**
- In Power BI, data loading is a process for importing data from multiple sources, so that relevant data can be used for analysis, visualization, and decision-making. Data loading is considered the first step in the process, because without any data, no analysis of decisions can be made. Data is the foundation stone for any analysis, as it helps in understanding the trends.  


**Question 2: Explain the difference between “Load” and “Transform Data” in Power BI.**
- In Power BI, while importing data, it gives two options: **load** and **transform data**. Load is used when we are sure that there are no irregularities in the data and the data is fit for the analysis.
- While transforming data is opted for when there are irregularities in the data that need to be removed, such as duplicate records, null values, or data type issues. Instead of loading data directly to Power BI, it takes to power query editor, where necessary changes can be made, and then it gets loaded for analysis and dashboard.


**Question 3: What are a Fact Table and a Dimension Table? Give examples from the dataset.**
- A fact table is the central table in a star or snowflake schema that stores quantitative data for analysis. It contains measurements, metrics, or facts about a business process. The fact table is characterized by its numeric measures and foreign keys that link to dimension tables.
- A dimension table provides descriptive context for the facts in the fact table. It contains attributes that describe the dimensions of the facts, such as product names, customer demographics, or time periods.


**Question 4: Why is the Star Schema preferred over the Snowflake Schema in Power BI?**
- In a star schema, there is only one fact table that is surrounded by various dimension tables. This structure makes the relationship between tables more understandable, reducing the number of joins.
- While in the snowflake schema dimension table has further relations with multiple tables, but this complicates the relationship as there are multiple tables for managing relationship, and requires multiple joins, which can slow down queries.
