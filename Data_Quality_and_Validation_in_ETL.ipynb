**Question 1: Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?**

- Data quality in the ETL process is about ensuring that data is accurate, complete, consistent, valid, and reliable before it is loaded into the target system. It is more than just data cleaning, as it includes validation of data, standardisation and monitoring throughout the ETL process instead of limiting it to correcting errors


**Question 2: Explain why poor data quality leads to misleading dashboards and incorrect decisions.**
- Poor quality of data can create problems in data analysis and decision-making, as all the decisions are made based on the results from data analysis. If the quality of data is poor, decisions will be made based on inaccurate data that may distort business metrics, resulting in poor strategic and operational decisions.

**Question 3: What is duplicate data? Explain three causes in ETL pipelines.**
- Duplicate data occurs when the same business record appears multiple times in a dataset. In ETL pipelines, duplicates can be caused by loading data from multiple sources, rerunning ETL jobs without proper controls, and missing business key validations during data integration.


**Question 4: Differentiate between exact, partial, and fuzzy duplicates.**
- Exact duplicates are records that are completely identical in all fields. Partial duplicates share some common attributes but differ in others. Fuzzy duplicates are records that represent the same entity but contain slight variations in spelling, formatting, or data entry.


**Question 5: Why should data validation be performed during transformation rather than after loading?**
- Data validation should be performed during transformation because errors can be detected and corrected before they reach the target database. This prevents bad data from being loaded, reduces rework, and improves the reliability of reports and analytics


**Question 6: Explain how business rules help in validating data accuracy. Give an example.**
- Business rules help validate data by ensuring it meets organisational requirements and standards. For example, a rule may require that every Customer_ID exists in the customer master table or that Txn_Amount cannot be NULL. Records violating these rules can be rejected or corrected before loading.


**Question 7: Question 7: Write an SQL query on Sales_Transactions to list all duplicate keys and their counts using the business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount )**
```
Select customer_ID, product_ID, txn_date, txn_amount,
Count(*) as duplicate_count from sales_transaction
Group by customer_id, product_id, txn_date, txn_amount
Having count(*) >1;
```


**Question 8: Identify Sales_Transactions.Customer_ID values that violate referential integrity when joined with Customers_Master and write a query to detect such violations.**
```
select distinct s.customer_id from sales_transactions s
left join customers_master c on s.customer_id = c.customerid
where c.customerid is null;
```
