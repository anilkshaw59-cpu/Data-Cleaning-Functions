**Question 1: What is Power BI, and why is it used in businesses?**

- Power BI is a business intelligence tool that is used by businesses as an analytics platform to convert raw data into actionable insights. It is used by businesses to present raw data from multiple sources in an interactive dashboard for quick, evidence-based decision-making.


**Question 2: Name and explain the three main components of Power BI.**

The three main components of Power BI are

- Power BI Desktop: This is the application where data is imported and reports are prepared.
- Power BI Services: This is used for publishing the dashboard prepared and share with people.
- Power BI App: This app is used for convenience, allowing reports and dashboards to be viewed on the go from mobile devices.


**Question 3: Explain the Power BI workflow.**

- The first step in the Power BI workflow is to **get data**, which can be sourced from multiple places such as a local host, databases, a data warehouse, cloud. These data are imported into Power BI with the help of connectors.
- The collected data may have some inconsistencies and irregularities that need to be cleaned and corrected. This is done in the transformation part in the Power Query Editor, where data is cleaned, duplicates are removed, null values are either replaced or removed, and a standard format is set.
- The next step in the workflow is to define the relationship. In a real-world situation, we need to deal with multiple tables, and data needs to be extracted from multiple tables to make a meaning full insights, thus establishing relationships between tables is important, which is part of data modelling.
- The next step is creating a dashboard for analytics with the help of data. Based on the business requirements, situation, and scenarios, business decisions can be made accordingly.
- This dashboard can be shared with clients and people through Power BI services.

**Question 6: Define Data View, Report View, and Model View. Explain the purpose of each view.**
- **Data View**: Under this view, all the data imported from various sources is shown in the form of a table.
- **Report View**: It is the area or canvas where all the visuals are created based on the analysis requirement. The purpose is to create charts, graphs, KPI. This is to prepare an interactive dashboard that can be used for decision-making.
- **Model View**: Provides information on the relationship shared by the tables, how the tables are connected, and what keys have been used. It defined the relationship as one-to-many, many-to-one, many-to-many, and one-to-one.
