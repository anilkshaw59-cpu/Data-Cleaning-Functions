```
Total Price = amazon_store_sales_2022[Quantity]*amazon_store_sales_2022[UnitPrice]
```

**Question 3: Create a measure Average Order Value (AOV).**

```
Average Order Value = AVERAGE(amazon_store_sales_2022[SalesAmount])
```

**Question 4: Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the OrderDate.**

```
YTD Sales =
CALCULATE(
    SUM(amazon_store_sales_2022[SalesAmount]),
    DATESYTD(amazon_store_sales_2022[OrderDate])
)
```

**Question 5: Using DAX, create a measure to rank each product according to its Total Sales, with the highest-selling product ranked first.,**

```
Rank =
RANKX(
    ALL(amazon_store_sales_2022[ProductCategory]),
    [Total Sales], , DESC,Dense)

```

**Question 8. Create a measure called "Date Difference" that calculates the number of days between the latest Order Date and today's date using the DATEDIFE function.**

```
Date Difference = datediff(max(amazon_store_sales_2022[OrderDate]),Today(), DAY)
```
