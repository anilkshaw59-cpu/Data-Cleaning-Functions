**Question 1: Define Data Transformation in ETL and explain why it is important.**

- Data transformation is part of the ETL process under which unorganised, inconsistent, raw or incomplete data is cleaned, standardised and made ready for analysis. The process includes the process of application of business rules, calculations, cleaning techniques and other methods.

- Data transformation is important to ensure the integrity of data; for example, outliers and duplicate entries within data may impact the overall results from the data. Data transformation plays a critical role in removing errors, duplicates and inconsistencies.


**Question 2: List any four common activities involved in Data Cleaning.**

Four common data cleaning activities are:
- 1.	Handling missing values
- 2.	Removing duplicate records
- 3.	Correcting inaccurate/invalid values
- 4.	Fixing inconsistent formatting


**Question 3: What is the difference between Normalisation and Standardisation?**

- Data normalisation is a process in which numerical values data is converted into a fixed range, usually from 0 to 1. It uses min-max scaling.  
- While data standardisation is based on mean and standard deviation, it also means setting a specific standard for data. For example, if the data is mentioned, setting dd-mm-yyyy as a standard format ensures all date values are mentioned in that format.


**Question 4: A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used.**

- One of the best techniques in this case will be to replace the missing values with the mean or median. The mean value will be preferred in case the data is normally distributed, while if the data contains outliers, the median is best suited for ignoring the impact of outliers.

- Another technique is to delete missing rows, but this can only be used when the portion of missing values is small.

**Question 7: Explain the difference between Data Integration and Data Mapping in ETL.**

- Data integration is about collecting data from multiple sources into a single place for analysis.
- Data mapping is about defining relationships between fields in different datasets. Example: Mapping Cust_ID in a CRM system to CustomerID in a sales database.


**Question 8: Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.**

- The reason why Z-score standardisation is preferred for scaling when there are outliers in the data is the use of mean and SD for scaling that minimises the impact of extreme values in the data.
- Under Minimum and maximum scaling, the minimum and maximum values within the data are used as a reference point; outliers can significantly affect these values.
- Therefore, Z-score Standardization provides a more reliable scaling method when datasets contain outliers.
