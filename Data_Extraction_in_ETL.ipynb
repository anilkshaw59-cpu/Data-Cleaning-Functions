**Question 1: Describe different types of data sources used in ETL with suitable examples.**

Four types of data sources are used in ETL: Files, Databases, APIs, and cloud sources.
- Flat Files are the simplest form of datasets that store data in the form of a table or plain text. For example, CSV files, Excel
- Databases are places where data is stored in structured, large-scale and transaction-based data. These types of sources include MySQL, SQL Server, and PostgreSQL, which are relational databases, while MongoDB and Cassandra are non-SQL databases.
- APIs are the sources that enable applications to communicate and exchange data in real time. For Example, Weather  API, it real time data related to temperature, humidity, and forecasts. Social media API such as Facebook and Instagram.
- Cloud-based storage or services that host large datasets and files. Such as Amazon S3, Google Cloud Storage, Azure Blob Storage.


**Question 2: What is data extraction? Explain its role in the ETL pipeline.**
- Data extraction is the first step in the ETL process where all the data from multiple sources is collected, such as files, databases, and online services. The role of extraction in the ETL pipeline is to pull the right data in the correct format from various sources for analytics purposes based on the requirement.


**Question 3: Explain the difference between CSV and Excel in terms of extraction and ETL usage.**

**CSV**
- CSV is known as comma separated value where data is stored in the form of text, with each line in the file representing a row, and the values of each row are separated by commas. This format is lightweight and easy to parse, which is why it is extensively used in ETL for quick extraction.

**EXCEL**
- Excel stores data in the form of a table; unlike CSV, multiple sheets can be stored in an Excel file along with formatting, formulas and charts. Because of these multiple features, the extraction process from Excel is more complex, as tools have to handle metadata, cell formatting, and sometimes embedded formulas.

**Question 4: Explain the steps involved in extracting data from a relational database.**

The first step in the process is to **establish a connection with the database**. Under this step tools such as MySQL Workbench, Python libraries is used to establish a link between the ETL system and the database server.

The next step in the process is **authentication of user**, under this step credentials like username, password database name is provided to ensure that data is only accessed by authorized users.

The third step is to **run SQL queries**, this is done to filter out only data that is needed using filters, joins, selecting specific columns, concatenating columns and others.

The last step is to extract the required data and loading the result. Extracted data is
- Exported into files (CSV, Excel), or
- Loaded directly into the ETL pipeline for transformation, or
- Sent to a data warehouse, reporting tool, or dashboard.

**Question 5: Explain three common challenges faced during data extraction.**
- **Missing and Inconsistent Formatting**: Data is extracted from multiple sources for example a marketing company may have fetch data form Google ads, Meta Ads, Digital Marketing and other sources. One of the major challenge is missing and inconsistent formatting of data from different sources. Every data sources has there own standards of data storage, aligning it as one data is a major challenge.
- **API Limit** : Many of the APIs restrict the number of request that can be processed per minute or per day, in this if the ETL process exceeds this limit, the API may block or delay responses.
- **Large File Size** Another major challenge is processing of large file, it slows down the loading and processing of data, consume more memory may cause extraction tools to crash.

**Question 7: Why are databases preferred for enterprise-level data extraction?**
- Databases are preferred for enterprise-level data extraction as they provide structured, reliable, and secure storage of large volumes of business data. At enterprise level there can be multiple points from where data is collected. Database helps in storing large amount of data and support SQL queries for precise extraction. Scalability is another  reason why it is preferred as  it can handle millions of records, security features like authentication and role-based access, and compatibility with ETL tools for automation


**Question 8: What steps should an ETL developer take when extracting data from large CSV files (1GB+)?**
- The best possible method for extracting data from large CSV file is to reading data in multiple parts instead of loading the whole file into memory. Another measure that can be taken is to using parallel/distributed tools (e.g., Spark, Dask) for faster processing.
