**Question 1: What is the purpose of a PivotTable in Excel?**

- Pivot table in excel is designed to summarise, analyse and evaluate large sets of data in best possible way. Pivot tables provide various features that can be used to analyse data from all angles to derive meaningful conclusion. It allows data visualisation with multiple filters, so that only data relevant for the decision can be extracted. The ability to control data visualisation makes identification of trends and patterns more easily accessible.

**Question 7: Why are Dashboards important for decision-making in business analysis?**
- Dashboards are important for decision making, because it provides all the information trends and patterns needed for decision-making at one place, that enables comparison of trends and patterns. Because of the dashboard we don’t have to go through tones of data and do manual analyse, it gets updated in real time as the data gets updated. One of the main feature of dashboard is its visualisation that allows for easy analyse of complex data.
