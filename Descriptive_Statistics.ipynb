**Question 1: Understanding Central Tendency (Easy)**

**A bakery tracks the daily muffin sales (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12]. What is the most representative value of their weekly sales, and why?**

- The average of daily sales is the most representative value for weekly sales, as it provides a central value for daily sales. In this case, the mean comes at 12.42. While the median is also 12, and the mode is also 12. Thus, the average is aligned closely around 12.42, which signifies that the mean is a reliable metric and provides an accurate representation.

**Q9: Combining Measures (Hard)**

**A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410]**

**Find the mean and standard deviation. What do these values together tell you about the family’s energy use pattern?**


Mean

    = (400+420+390+450+410)/5
    = 414
This means that the average electrical consumption of the family is 414 kWh

Variance

    = ((400-414)^2 + 420-414)^2 + 390-414)^2 + (450-414)^2 + (410-414)^2) / 5
    = (142 + 62 + 242 + 362 + 42) / 5
    = (196+ 36+ 576+ 1296+ 16) / 5
    = 2120/ 5
    = 424

Standard Deviation

    = SquareRoot(424)
    = 20.59

- The SD of 20.59 signifies that the variation in electrical consumption is about ±20 kWh, the consumption is stable and doesn’t fluctuate wildly. This makes energy consumption prediction easy.


**Q10: Practical Application (Hard)**

**A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21]. Find the mean, median, mode, range, and standard deviation. What insights can these measures provide about the player’s scoring performance?**


Mean

    = (15+ 18+20+22+25+17+19+21)/ 8
    = 19.625

Median

Arranging the data: 15, 17, 18, 19, 20, 21, 22, 25


    Median = (4th + 5th value ) / 2
    = (19+20) / 2
    = 19.5

Mode

- In this case, there is no mode as all the values occur once and are of equal frequency.

Range

    = Maximum – Minimum
    = 25- 15
    = 10

Variance

    = ((15-19.625)^2 + (17-19.625)^2 + (18-19.625)^2 + (19-19.625)^2 + (20-19.625)^2 + (21-19.625)^2 + (22-19.625)^2 + (25-19.625)^2) / 8

    = 67.875 / 8
    = 8.48

Standard Deviation

    = Squareroot(8.48)
    = 2.91

- Overall performance of players is stable, as the mean is 19.625 while the median is also 19.5. This shows that the data is not affected by outliers. SD of 2.91 shows that the deviation will be ±2.91. The range is 10, which occurs occasionally, but the overall performance is stable
