**Question 4: What is the concept of standardisation (z-score), and why is it important in data analysis? Explain the formula and how standardisation transforms a dataset.**

- Standardisation, or (z-score), is a process where data points in a dataset are transformed so that their deviation from the mean can be determined. It is a process that makes comparison of different sets of data easier, as the units of all the values are standardised on a single scale. It is important in data analysis as it makes datasets comparable and highlights outliers in the dataset.

- It’s important because it makes datasets comparable, highlights outliers, and allows probabilities to be calculated using the standard normal distribution.
    
    The formula is (𝑥 -  𝜇)/ 𝜎
    
    𝑥 = raw data point

    𝜇 = mean of the dataset

    𝜎 = standard deviation

- Suppose the data point is 20 and the mean is 30, while SD is 10, then the z score will be 20-30/10 = -1

- This will be pointed to the left of the mean.



**Question 5: What is Kurtosis and its type?**
- Kurtosis is used for measuring the heaviness or lightness of tails in a distribution when compared to the normal distribution. It is used to determine the frequency of extreme values or outliers.  There are three types of Kurtosis.

- Mesokurtic, in which the tails of the distribution are similar to the normal distribution
- Platykurtic, where the tail of the distribution remains thin with fewer extreme values
- Leptokurtic, where the frequency of extreme values is very high


**Question 6: Explain why the uniform distribution is a good model for the outcome of rolling a fair die.**

- The reason why uniform distribution is good model for the outcome of rolling a fair die is because of the equal probability of six outcomes that is 1/6. Uniform distribution is a good model as it reflects the fairness of the die and ensures no bias toward any number, and also captures the equal likelihood of all possible results.

**Question 8: How does log transformation help in stabilising variance and making data more normally distributed?**

- Log transformation reduces skewness and stabilises variance by compressing large values and stretching small ones. This makes data less heteroscedastic and often more symmetric, helping it approximate a normal distribution for statistical analysis.
